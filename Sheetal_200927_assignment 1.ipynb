```python
for i in range(0,len(df)):
   if ',' in str(df.iloc[i,8]):
        cl=df.iloc[i,8].split(',')
        df.iloc[i,8]=int(cl[0])
        df.iloc[i,11]=int(cl[1])
        if len(cl)>2:
            df.iloc[i,12]=int(cl[2])
   elif int(df.iloc[i,8])>100:
            df.iloc[i,8]=int(df.iloc[i,8])%100
            df.iloc[i,8]=int(df.iloc[i,8])//100
```
